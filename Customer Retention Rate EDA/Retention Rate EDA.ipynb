```python
# Ensure 'Order Date' is in datetime format
#errors='coerce' means, if a string that doesn’t match the specified format, it will replace that value with NaT (Not a Time)
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%m/%d/%y', errors='coerce')
```

```python
# Extract the year and month from 'Order Date'
data['Order Year-Month'] = data['Order Date'].dt.to_period('M')
```

```python
# Group the data by 'Customer ID' and 'Order Year-Month'
customer_purchases_monthly = data.groupby(['Customer ID', 'Order Year-Month']).size().reset_index(name='Purchase Count')
```

```python
# Pivot Table
customer_pivot_monthly = customer_purchases_monthly.pivot(index='Customer ID', columns='Order Year-Month', values='Purchase Count')
```

```python
# Calculate retention by checking how many customers from one month made a purchase in the next month
retention_data_monthly = {}
months = customer_pivot_monthly.columns.tolist()

for i in range(len(months) - 1):
    current_month = months[i]
    next_month = months[i + 1]
    
    customers_current_month = customer_pivot_monthly[current_month].dropna().index
    customers_next_month = customer_pivot_monthly[next_month].dropna().index
    
    retained_customers = customers_current_month.intersection(customers_next_month)
    retention_rate_monthly = len(retained_customers) / len(customers_current_month) if len(customers_current_month) > 0 else 0
    
    retention_data_monthly[f'{current_month} -> {next_month}'] = round(retention_rate_monthly * 100, 2)
```

```python
# Convert retention data to a DataFrame for better visualization
retention_df_monthly = pd.DataFrame(list(retention_data_monthly.items()), columns=['Month Transition', 'Retention Rate (%)'])
```

```python
# Display the retention data for month-wise retention
retention_df_monthly
```

```python
# Sorting 'Retention Rate (%)' in descending order
top_10_retention = retention_df_monthly.sort_values(by='Retention Rate (%)', ascending=False).head(10)

# Display the first 10 result (maximum)
top_10_retention
```
